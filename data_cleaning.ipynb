## Country Wise Data

```python
def country_wise_clean_data(file_name):
    
    # read csv file
    df = pd.read_csv(file_name, skiprows=1)
    # replace 'No data' with 'na' inside the dataframe
    df = df.replace('No data', 'na')
    # convert to long format
    df = df.melt(id_vars='Country', value_vars=df.columns[1:], var_name='Year', value_name='Count')

    # print('Raw Data :\n', '='*20, '\n', df.head())

    # clean the data
    for col in ['Count']:
        df[col] = df[col].str.replace('&lt;', '<')
        df[col] = df[col].str.replace('&gt;', '>')
        df[col] = df[col].str.replace('<', '')
        df[col] = df[col].str.replace('>', '')    
        df[col] = df[col].str.strip()
        df[col] = df[col].str.replace(' ', '')
        df[col+'_median'] = df[col].str.extract('([0-9.]+)')
        df[col+'_min'] = df[col].str.extract('[0-9.]+\[([0-9.]+)')
        df[col+'_max'] = df[col].str.extract('[0-9.]+\[[0-9.]+\–([0-9.]+)')
        
    print('\nClean Data :\n', '='*20, '\n', df.head())

    # add 'WHO Region' column
    df['WHO Region'] = df['Country'].map(who_region)
    # print(df[df['WHO Region'].isna()]['Country'].unique())
    
    # save the file
    save_as = file_name[:-4]+'_clean.csv'
    df.to_csv(save_as, index=False)
```

## Prevention of MTC

```python
def prev_mtc_clean_data(file_name):
    
    df = pd.read_csv(file_name)
    df.columns = ['Country', 'Received Antiretrovirals', 'Needing antiretrovirals', 'Percentage Recieved']
    df = df.drop(0)
    
#     print('Raw Data :\n', '='*20, '\n', df.head())

    for col in df.columns[2:]:
        df[col] = df[col].str.replace('&lt;', '<')
        df[col] = df[col].str.replace('&gt;', '>')
        df[col] = df[col].str.replace('<', '')
        df[col] = df[col].str.replace('>', '')    
        df[col] = df[col].str.strip()
        df[col] = df[col].str.replace(' ', '')
        df[col+'_median'] = df[col].str.extract('([0-9.]+)')
        df[col+'_min'] = df[col].str.extract('[0-9.]+\[([0-9.]+)')
        df[col+'_max'] = df[col].str.extract('[0-9.]+\[[0-9.]+\–([0-9.]+)')
    
    # add 'WHO Region' column
    df['WHO Region'] = df['Country'].map(who_region)
    
    print('\nClean Data :\n', '='*20, '\n', df.head())

    save_as = file_name[:-4]+'_clean.csv'
    df.to_csv(save_as, index=False)
```

## ART

```python
df = pd.read_csv('./who/art_coverage_by_country.csv')
df = df.iloc[1:, :4]
df.columns = ['Country', 'Estimated ART coverage among people living with HIV (%)', 
              'Reported number of people receiving ART', 
              'Estimated number of people living with HIV']
df = df[['Country', 'Reported number of people receiving ART', 
         'Estimated number of people living with HIV',
         'Estimated ART coverage among people living with HIV (%)']]
df['Reported number of people receiving ART'] = df['Reported number of people receiving ART'].str.replace(' ', '')

for col in df.columns[2:]:
    df[col] = df[col].str.replace('&lt;', '<')
    df[col] = df[col].str.replace('&gt;', '>')
    df[col] = df[col].str.replace('<', '')
    df[col] = df[col].str.replace('>', '')    
    df[col] = df[col].str.strip()
    df[col] = df[col].str.replace(' ', '')
    df[col+'_median'] = df[col].str.extract('([0-9.]+)')
    df[col+'_min'] = df[col].str.extract('[0-9.]+\[([0-9.]+)')
    df[col+'_max'] = df[col].str.extract('[0-9.]+\[[0-9.]+\–([0-9.]+)')
        
df['WHO Region'] = df['Country'].map(who_region)
# print(df[df['WHO Region'].isna()]['Country'].unique())
    
df.to_csv('./who/art_coverage_by_country_clean.csv', index=False)
df.head()
```
